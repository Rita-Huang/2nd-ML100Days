 * **計算任意欄位的平均數及標準差**
 * **畫出任意欄位的直方圖**

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/Part01'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

* **計算欄位"AMT_INCOME_TOTAL"的平均數及標準差並繪出其直方圖**

```python
AIT = app_train["AMT_INCOME_TOTAL"]
print("Mean : {:.2f}".format(AIT.mean()))
print("Standard Deviation : {:.2f}".format(AIT.std()))
```

```python
plt.hist(AIT.head(1000), bins=40, rwidth=0.8)
plt.grid(color='g',linestyle='--', linewidth=1,alpha=0.4)
plt.xlabel("INCOME")
plt.ylabel("Number of Clients")
plt.show()
```

* **繪出欄位"WALLSMATERIAL_MODE"的直方圖**

```python
HM = app_train["WALLSMATERIAL_MODE"].dropna()
print(HM.unique())
```

```python
plt.hist(HM, bins=20, rwidth=0.9)
plt.grid(color='g',linestyle='--', linewidth=1,alpha=0.4)
plt.xlabel("Walls Material")
plt.ylabel("Couns")
plt.show()
```
